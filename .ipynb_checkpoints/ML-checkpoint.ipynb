### Explore databases using seaborn pairplot and sklearn LinearRegression models to find any correlations between features. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing, svm
```

```python
full_rest = pd.read_csv('Resources/full_rest_data_new.csv')
```

```python
full_rest
```

```python
#create inputs and target variables
features = ['Pop', 'PCI', 'price', 'rating', 'review']
X = full_rest[features]
y = full_rest['Total'].values.reshape(-1, 1)

#import Standard Scaler 
from sklearn.preprocessing import StandardScaler

#Scale inputs and split into training and testing sets
sc_X = StandardScaler()
X = sc_X.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

model = LinearRegression()

#fit the model and run
model.fit(X_train, y_train)
training_score = model.score(X_train, y_train)
testing_score = model.score(X_test, y_test)
print(training_score)
print(testing_score)
coef = model.coef_
print(model.intercept_)
```

```python
#create coeffecient dataframe
coef_df = pd.DataFrame(coef)
```

```python
coef_df['Intercept'] = 232.28656073
features = ['Pop', 'PCI', 'price', 'rating', 'review', 'Intercept']
```

```python
#plot coeffecients
plt.ylim(ymax = 500, ymin = -10)
plt.bar(features, coef_df.iloc[0])
```

```python
#create array of predicted restaurant totals for cities
print(X)
predicted_scaled = model.predict(X)
```

```python
#create pandas dataframe from array and merge with full restaurant dataframe
predicted_scaled_df = pd.DataFrame(predicted_scaled)
rest_predicted_scaled = full_rest.merge(predicted_scaled_df, left_index=True, right_index=True)
rest_predicted_scaled = rest_predicted_scaled.rename(columns={0:'Predicted'})

#Create a column that contains prediction Errors
rest_predicted_scaled['Error'] = rest_predicted_scaled['Predicted'] - rest_predicted_scaled['Total']
```
