```python
#drop columns
pop.drop(['SUMLEV', 'STATE', 'COUNTY', 'PLACE', 'COUSUB', 'CONCIT',
       'PRIMGEO_FLAG', 'FUNCSTAT','CENSUS2010POP',
       'ESTIMATESBASE2010', 'POPESTIMATE2010', 'POPESTIMATE2011',
       'POPESTIMATE2012', 'POPESTIMATE2013', 'POPESTIMATE2014',
       'POPESTIMATE2015', 'POPESTIMATE2016', 'POPESTIMATE2017'], axis=1, inplace=True)
```

```python
pop = pop.rename(columns={'NAME':'City', "STNAME":'State'})
```

```python
PCI = pd.read_csv('PCI.csv')
```

```python
PCI
```

```python
st_abbr = pd.read_csv('st_abbr.csv')
```

```python
st_abbr.drop(['Abbrev'], axis=1, inplace=True)
```

```python
#seperate city and state into seperate columns
PCI['City'] = PCI['GeoName'].apply(lambda x: x.partition(',')[0])
PCI['State'] = PCI['GeoName'].apply(lambda x: x.partition(',')[2])
```

```python
#drop now useless GeoName and TimePeriod columns
PCI.drop(['GeoName', 'TimePeriod'], axis=1, inplace=True)
```

```python
PCI.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
#strip empty characters from columns
PCI['State'] = PCI['State'].apply(lambda x: x.strip())
PCI['City'] = PCI['City'].apply(lambda x: x.strip())
```
